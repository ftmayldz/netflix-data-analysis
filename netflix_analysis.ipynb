```python
#Goal of project :
# ✅ Analyze user viewing habits
# ✅ See which genres/content are more popular
# ✅ Examine how variables such as country, year, and genre affect content distribution
# ✅ Generate useful insights from the data
# ✅ Demonstrate practical skills in Python + pandas + visualization



import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 

df=pd.read_csv("netflix.csv")

df.head()
```

```python
#(EDA)

df.info()
df.isnull().sum()
```

```python
df.describe(include='all')
#Dataset contains missing values in several categorical columns (director ,cast,country) , therefore data cleaning and preprocessing are required before advanced analysis 
```

```python
#Cleaning 
df['date_added']=pd.to_datetime(df['date_added'],errors='coerce')
df['year_added']=df['date_added'].dt.year

df=df.dropna(subset=['country'])
```

```python
#First graph - Film vs TV 
df['type'].value_counts().plot(kind='bar')
plt.title("Movies vs TV Shows")
plt.show()


```

```python
#Content by countries (Top 10)
df['country'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Countries")
plt.show()
```

```python
#Content increase over to years 
df.groupby('year_added').size().plot()
plt.title("Content Added Over Years")
plt.show()
```

```python
#Genre analysis 
df['listed_in'].str.split(',').explode().str.strip().value_counts().head(10).plot(kind='bar')
plt.title("Top Genres")
plt.ylabel("Count")
plt.show()
```
